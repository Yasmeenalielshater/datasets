```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("C:/Users/XPRISTO/Desktop/UdemyCoursesDataset.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
for x in df.index:
    if(df.loc[x, "num_lectures"]==0):
        df.drop(x,inplace=True)
```

```python
df['course_id'].nunique()
```

```python
df.drop(columns=['course_id'],inplace=True)
```

```python
df.shape
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df['year'] = df['published_timestamp'].dt.year
```

```python

df['year'].value_counts().sort_index()
```

```python
df
```

```python
df['month']=df['published_timestamp'].dt.month
```

```python
numerical_df = df.select_dtypes(exclude='object')


plt.figure(figsize=(10, 10))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap for Numerical Features")
plt.show()
```

```python
sns.countplot(x='is_paid',data=df)
plt.title('paid and free courses')
plt.show()
```

```python
plt.pie(
    df['is_paid'].value_counts(),
    labels=('Paid', 'Free'),
    autopct='%1.2f%%',
    shadow=True,
    explode=[0, 0.15]  
)

plt.title('Paid and Free Courses')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=df,
    x='num_subscribers',
    y='subject',
    estimator='mean',  # متوسط
    order=df.groupby('subject')['num_subscribers'].mean().sort_values(ascending=False).index,
    palette='mako'
)
plt.title('Average Number of Subscribers per Subject')
plt.xlabel('Average Subscribers')
plt.ylabel('Subject')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Average Demand (Free vs Paid)')
sns.barplot(x='is_paid', y='num_subscribers', data=df, palette='pastel')
plt.xlabel("Is Paid (True = Paid, False = Free)")
plt.ylabel("Average Number of Subscribers")
plt.show() 
```

```python
df["subject"].value_counts()
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    df['subject'].value_counts(), 
    labels=df['subject'].value_counts().index,  
    autopct='%1.2f%%',                      
    shadow=True, 
    explode=[0.1, 0, 0, 0]                      
)
plt.title('Number of Courses in Each Subject')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='subject',
    y='num_subscribers',
    hue='is_paid',
    data=df,
    estimator='mean',
    order=df.groupby('subject')['num_subscribers'].mean().sort_values(ascending=False).index,
    palette="Set2"
)
plt.title('Average Demand by Course Subject', fontsize=14)
plt.xlabel('Course Subject')
plt.ylabel('Average Number of Subscribers')
plt.xticks(rotation=45)    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    y="level",  
    order=df.level.value_counts().index,
    color='#4682B4' 
)
plt.title('Number of Courses by Level', fontsize=14)
plt.xlabel('Number of Courses')
plt.ylabel('Course Level')
plt.tight_layout()
plt.show()

```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.dropna(subset=['price'], inplace=True)

```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='price', bins=30,  color='#4682B4' , edgecolor='black')

plt.title('Distribution of Course Prices', fontsize=16)
plt.xlabel('Price ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)


plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='price',y='num_subscribers',data=df)
plt.show()
```

```python
sns.scatterplot(x="num_lectures",y="num_subscribers",data=df)
plt.show()
```

```python

df['price'] = pd.to_numeric(df['price'], errors='coerce')


plt.figure(figsize=(10,6))
plt.plot(df.groupby('year')['price'].sum(), linestyle='-', linewidth=3, alpha=0.85, color='blue')

plt.title("Total Price of Courses Over the Years", fontsize=15)
plt.xlabel('Years', color='k', fontsize=12)
plt.ylabel('Total Price ($)', color='k', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='month')
plt.title('distribuation of courses publiched over monthes')
plt.xlabel('month')
plt.ylabel('number of courses')
plt.xticks(range(0,12),{'Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'})
plt.show()
```

```python
  
z = df.sort_values(by='num_subscribers', ascending=False).head(5)


sns.catplot(data=z, kind='bar', height=8, aspect=2.4, 
            x='course_title', y='num_subscribers', palette='Blues_d')

plt.xticks(rotation=45, ha='right')   
plt.title('Top 5 Courses by Number of Subscribers')
plt.tight_layout()
plt.show()
```

```python
z = df.sort_values(by='num_reviews', ascending=False).head(5)


sns.catplot(data=z, kind='bar', height=8, aspect=2.4, 
            x='course_title', y='num_reviews', palette='Blues_d')

plt.xticks(rotation=45, ha='right')   
plt.title('Top 5 Courses by Number of reviews')
plt.tight_layout()
plt.show()
```

```python

df['price'] = pd.to_numeric(df['price'], errors='coerce')   
df['price'].fillna(0, inplace=True)    
plt.figure(figsize=(14,6))
sns.histplot(data=df, x='price', bins=40, palette='Blues_d', edgecolor='black')

plt.title('Distribution of Course Prices', fontsize=16)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
